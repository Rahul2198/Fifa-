```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('data_cleaned.csv')
df.head()
```

```python
plt.figure(figsize=(50,40))

sns.heatmap(df.corr(), annot=True)
```

```python
cols = ['Age','Overall','Potential','Acceleration','SprintSpeed',"Agility","Stamina",'Strength','Preferred Foot']
df_small = df[cols]
```

```python

sns.pairplot(df_small, hue ='Preferred Foot',palette=["black", "red"],plot_kws=dict(s=50, alpha =0.8),markers=['^','v'])
```

```python
sns.distplot(df['Age'],kde=True,bins=30)
```

```python
sns.countplot(x="Body Type",hue="Preferred Foot", data=df)
```

```python
f, ax = plt.subplots(figsize=(8, 6))
sns.countplot(x="Preferred Foot", data=df, color="c")
plt.show()
```

```python
df['International Reputation'].value_counts()
```

```python
sns.boxplot(x="International Reputation", y="Potential", data=df)
```

```python
sns.pointplot(x="International Reputation", y="Potential", data=df)
```

```python
plt.figure(figsize=(15,10))

sns.pointplot(x="International Reputation", y="Potential", hue="Preferred Foot", 
              data=df, markers=["o", "x"], linestyles=["-", "--"])
```

```python
f, ax = plt.subplots(figsize=(8, 6))
sns.barplot(x="International Reputation", y="Potential", hue="Preferred Foot", data=df)
plt.show()
```

```python
 sns.lineplot(x="Stamina", y="Strength", data=df)
```

```python
sns.regplot(x="International Reputation", y="Potential", data=df, x_jitter=.01)
```
